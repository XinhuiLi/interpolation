```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# SZ: layer 3, seed 3
res_path = "/data/users4/xli/interpolation/results/sfnc_sz/vae/hypopt"
dict_corr_train_sz, dict_corr_test_sz = {}, {}

for layer in [2,3,5,7]:
    dict_corr_train_sz[f"{layer}"] = []
    dict_corr_test_sz[f"{layer}"] = []
    for seed in range(10):
        res_file_train = f"{res_path}/layer{layer}/seed{seed}/corr.npy"
        corr_train = np.load(res_file_train)
        corr_train_valid = corr_train[~np.isnan(corr_train)]
        dict_corr_train_sz[f"{layer}"] += list(corr_train_valid)
        res_file_test = f"{res_path}/layer{layer}/seed{seed}/corr_test.npy"
        corr_test = np.load(res_file_test)
        corr_test_valid = corr_test[~np.isnan(corr_test)]
        dict_corr_test_sz[f"{layer}"] += list(corr_test_valid)
        if layer == 3:
            print(f"seed {seed} training performance: {np.mean(corr_train_valid):.4f}+/-{np.std(corr_train_valid):.4f}; inference performance: {np.mean(corr_test_valid):.4f}+/-{np.std(corr_test_valid):.4f}")
    print(f"layer {layer} training performance: {np.mean(dict_corr_train_sz[f"{layer}"]):.4f}+/-{np.std(dict_corr_train_sz[f"{layer}"]):.4f}; inference performance: {np.mean(dict_corr_test_sz[f"{layer}"]):.4f}+/-{np.std(dict_corr_test_sz[f"{layer}"]):.4f}")
```

```python
# ASD: layer 2 (64, 16), seed 6
res_path = "/data/users4/xli/interpolation/results/sfnc_asd/vae/hypopt"
dict_corr_train_asd, dict_corr_test_asd = {}, {}

for layer in [2,3,5,7]:
    dict_corr_train_asd[f"{layer}"] = []
    dict_corr_test_asd[f"{layer}"] = []
    for seed in range(10):
        res_file_train = f"{res_path}/layer{layer}/seed{seed}/corr.npy"
        corr_train = np.load(res_file_train)
        corr_train_valid = corr_train[~np.isnan(corr_train)]
        dict_corr_train_asd[f"{layer}"] += list(corr_train_valid)
        res_file_test = f"{res_path}/layer{layer}/seed{seed}/corr_test.npy"
        corr_test = np.load(res_file_test)
        corr_test_valid = corr_test[~np.isnan(corr_test)]
        dict_corr_test_asd[f"{layer}"] += list(corr_test_valid)
        if layer == 2:
            print(f"seed {seed} training performance: {np.mean(corr_train_valid):.4f}+/-{np.std(corr_train_valid):.4f}; inference performance: {np.mean(corr_test_valid):.4f}+/-{np.std(corr_test_valid):.4f}")
    print(f"layer {layer} training performance: {np.mean(dict_corr_train_asd[f"{layer}"]):.4f}+/-{np.std(dict_corr_train_asd[f"{layer}"]):.4f}; inference performance: {np.mean(dict_corr_test_asd[f"{layer}"]):.4f}+/-{np.std(dict_corr_test_asd[f"{layer}"]):.4f}")
```

```python
def plot_boxviolin(df, ax, title):
    sns.set_style("whitegrid")
    sns.violinplot(data=df, fill=False, gap=.1, ax=ax, legend=False)
    ax.set_xticklabels([2,3,5,7], fontsize=10)
    ax.set_xlabel("Number of layers", fontsize=11)
    ax.set_ylabel("Pearson correlation", fontsize=11)
    ax.set_title(title, fontweight='bold', fontsize=13)
    ax.set_ylim([0.4,1.05])
    for i, k in enumerate(df.keys()):
        ax.text(i, 1, f"{np.mean(df[k]):.4f}", ha='center', va='bottom', color='k')
        ax.text(i, 0.96, f"({np.std(df[k]):.3f})", ha='center', va='bottom', color='k')

fig, axes = plt.subplots(2, 2, figsize=(9, 6.4))
plot_boxviolin(dict_corr_train_sz, axes[0,0], "FBIRN sFNC (training)")
plot_boxviolin(dict_corr_test_sz, axes[1,0], "FBIRN sFNC (test)")
plot_boxviolin(dict_corr_train_asd, axes[0,1], "ABIDE I sFNC (training)")
plot_boxviolin(dict_corr_test_asd, axes[1,1], "ABIDE I sFNC (test)")
plt.tight_layout()
plt.savefig('hypopt_sfnc_vae.pdf')
```
