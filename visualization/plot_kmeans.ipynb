```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
res_path_fbirn = "/data/users4/xli/interpolation/results/dfnc_sz/vae/hypopt/layer7/seed1"
loss_fbirn = np.load(os.path.join(res_path_fbirn, "loss_l2_list.npy"))
vae_loss_fbirn = np.load(os.path.join(res_path_fbirn, "vae_loss_l2_list.npy"))

res_path_abide = "/data/users4/xli/interpolation/results/dfnc_asd/vae/hypopt/layer5/seed7"
loss_abide = np.load(os.path.join(res_path_abide, "loss_l2_list.npy"))
vae_loss_abide = np.load(os.path.join(res_path_abide, "vae_loss_l2_list.npy"))
```

```python
label_fontsize = 11
title_fontsize = 13
loss_list = [[vae_loss_fbirn, loss_fbirn], [vae_loss_abide, loss_abide]]
title_list = [["a  FBIRN generated clustering", "b  FBIRN original clustering"], ["c  ABIDE I generated clustering", "d  ABIDE I original clustering"]]

fig, axes = plt.subplots(2,2,figsize=(8,5.5))

for i in range(2):
    for j in range(2):
        ax = axes[i,j]
        ax.plot(range(2,10), loss_list[i][j])
        ax.set_xlabel("$k$", fontsize=label_fontsize)
        ax.set_ylabel("$L_2$ loss", fontsize=label_fontsize)
        ax.set_title(title_list[i][j], fontweight='bold', fontsize=title_fontsize)
        ax.set_xticks(range(2,10))
        ax.grid(True)

plt.tight_layout()
plt.savefig('kmeans_loss.pdf')
```
